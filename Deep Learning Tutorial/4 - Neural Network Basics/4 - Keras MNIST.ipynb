```python
# import the necessary packages
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from keras.models import Sequential
from keras.layers.core import Dense
from keras.optimizers import SGD
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline
sns.set()
```

```python
# grab the MNIST dataset (if this is your first time running this
# script, the download may take a minute -- the 55MB MNIST dataset
# will be downloaded)
print("Loading MNIST (full) dataset...")
dataset = datasets.fetch_mldata("MNIST Original")
print("Done!")

# scale the raw pixel intensities to the range [0, 1.0], then
# construct the training and testing splits
data = dataset.data.astype("float") / 255.0
(trainX, testX, trainY, testY) = train_test_split(data, dataset.target, test_size=0.25)
```

```python
# convert the labels from integers to vectors
# basically performing one-hot-encoding
lb = LabelBinarizer()
trainY = lb.fit_transform(trainY)
testY = lb.transform(testY)

print(trainY[0])
```

```python
# define the 784-256-128-10 architecture using Keras
# feed-fowards network (Sequential)
model = Sequential()
# first layer using 784 inputs (28x28) and learns 256 outputs, using the sigmoid activation function
model.add(Dense(256, input_shape=(784,), activation="sigmoid"))
# second layer learns 128 outputs, using the sigmoid activation function
model.add(Dense(128, activation="sigmoid"))
# final layer learns 10 weights for each of the 10 classes (0-9). It uses the softmax to get probabilities
model.add(Dense(10, activation="softmax"))
```

```python
# train the model usign SGD
print("[INFO] training network...")
# using SGD optimizer with a learning rate of 0.01
sgd = SGD(0.01)
# using cross-entropy loss function (needs vector labels instead of integer labels)
model.compile(loss="categorical_crossentropy", optimizer=sgd, metrics=["accuracy"])
# now we train our network passing both a training set and a validation set
# we should actually have a separated test set but this is just a quick example
H = model.fit(trainX, trainY, validation_data=(testX, testY), epochs=100, batch_size=128)
```

```python
# evaluate the network
print("[INFO] evaluating network...")
predictions = model.predict(testX, batch_size=128)
truth = testY.argmax(axis=1)
pred = predictions.argmax(axis=1)
class_names = lb.classes_
c_report = classification_report(truth, pred, class_names )
print(c_report)
```

```python
# plot the training loss and accuracy
plt.figure(figsize=(12,6))
plt.plot(np.arange(0, 100), H.history["loss"], label="train_loss")
plt.plot(np.arange(0, 100), H.history["val_loss"], label="val_loss")
plt.plot(np.arange(0, 100), H.history["acc"], label="train_acc")
plt.plot(np.arange(0, 100), H.history["val_acc"], label="val_acc")
plt.title("Training Loss and Accuracy")
plt.xlabel("Epoch #")
plt.ylabel("Loss/Accuracy")
plt.legend();
```

We get around 92% accuracy on the full MNIST dataset (last example with our own Neural Network implementation focused on a subset of the original dataset with data simplified to 8x8 digits). The full dataset is much more complicated with 70000 images encoded as 28x28 pixels. While this is good we can do much better with CNNs.

We'll see that this simple type of networks does barely better than random guessing on more complicated datasets.

The MNIST dataset is the very definition of a benchmark dataset where it's very easy to achieve very high values of accuracy, and is not representative of the images we'll see in the real world.
