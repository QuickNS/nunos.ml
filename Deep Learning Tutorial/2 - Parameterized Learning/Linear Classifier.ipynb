## Score Functions

This is a cooked up example just to show the process of how a scoring function works to predict a label value for an image.
It's the job of the traning phase to come up with the W and b vectors that minimize error (loss function). The scoring function just multiplies the image features by the weights and adds the bias to achieve a conclusion, which is depicted here as a score value for every possible class.

```python
# import the required packages
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# initialize class labels
labels = ['dog' ,'cat', 'panda']

# set seed of the random number generator so we can reproduce results
np.random.seed(1)
```

```python
# randomly initialize our weight matrix and bias vector -- in a
# *real* training and classification task, these parameters would
# be *learned* by our model, but for the sake of this example,
# let's use random values
W = np.random.randn(3, 3072)
b = np.random.randn(3)

print(W.shape)
print(b.shape)
```

```python
# load our example image, resize it, and then flatten it into our
# "feature vector" representation
orig = cv2.imread("beagle.png")
image = cv2.resize(orig, (32, 32)).flatten()
```

```python
# compute the output scores by taking the dot product between the
# weight matrix and image pixels, followed by adding in the bias
scores = W.dot(image) + b
```

```python
# loop over the scores + labels and display them
for (label, score) in zip(labels, scores):
    print("[INFO] {}: {:.2f}".format(label, score))
```

```python
# draw the label with the highest score on the image as our
# prediction
cv2.putText(orig, "Label: {}".format(labels[np.argmax(scores)]),
    (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

# display our input image
plt.axis("off")
plt.imshow(cv2.cvtColor(orig, cv2.COLOR_BGR2RGB));
```

So, we see our model is basically defined by the weight matrix and the bias vector. This representation is very compact and quite powerful, as well as very fast to predict values against unseen data

## What about loss functions?

```python
# Multi-class SVM loss
def hingeLoss(scores, correct_class):
    # for each wrong class get the max between 0 and the difference of scoring values of each class and the right class
    loss = 0
    for i in range(0,len(scores)):
        if i != correct_class:
            loss = loss + max(0, scores[i] - scores[correct_class] + 1)
    return loss

# correct class is dog (0)
loss_dog = hingeLoss(scores, 0)
    
print(loss_dog)
```

As the loss function is 0, we know this is a correct prediction, so the weights and biases wouldn't be further adjusted in training. If, however, the loss function is > 0, then the learning would probably continue (eventually it needs to stop and 0 is not a realistic goal).  
The objective is to minimize the loss function which would be the mean value of the loss function applied to all datapoints.

Let's try adding a new image, and this one is incorrectly classified by our scoring function.

```python
# let's bring in an image of a panda
orig = cv2.imread("panda.jpg")
image = cv2.resize(orig, (32, 32)).flatten()

# score it
scores = W.dot(image) + b

# loop over the scores + labels and display them
for (label, score) in zip(labels, scores):
    print("[INFO] {}: {:.2f}".format(label, score))
    
# draw the label with the highest score on the image as our prediction
cv2.putText(orig, "Label: {}".format(labels[np.argmax(scores)]),
    (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

# display our input image
plt.axis("off")
plt.imshow(cv2.cvtColor(orig, cv2.COLOR_BGR2RGB));
```

```python
# apply loss function
# correct class is 2
loss_panda = hingeLoss(scores, 2)
    
print(loss_panda)
```

```python
# loss function result (mean)
(loss_dog + loss_panda) / 2
```

The objective is to run an optimization process to minimize this value throughout the training phase. Optimization mechanisms such as gradient descent can be used to adjust the weight matrix and bias vector to achieve this goal. Hinge-loss explained above is useful, however in deep learning we usually deel with **softmax** classifiers that output probabilities, in which case a better approach to a loss function is **cross-entropy**

```python
def crossEntropyLoss(scores, correct_class):
    import numpy as np
    scores = np.abs(np.exp(scores))
    scores = scores / scores.sum()
    loss = - np.log(scores[correct_class])
    print("Confidence level for correct prediction: %.2f%%" % (scores[correct_class] * 100))
    return loss

# adding more reasonable values to what a softmax classifier would output
scores = [-3.44, 1.16, 3.91]
print("Loss function value: %.5f" % crossEntropyLoss(scores, 2))
```
