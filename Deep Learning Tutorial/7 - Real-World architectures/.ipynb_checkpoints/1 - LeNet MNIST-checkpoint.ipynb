## LeNet Architecture

INPUT => CONV => RELU => POOL => CONV => RELU => POOL => FC => RELU => FC

```python
# import the necessary packages
from ndl.nn.conv import LeNet
from keras.optimizers import SGD
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn import datasets
from keras import backend as K
import matplotlib.pyplot as plt
import numpy as np
```

```python
# grab the MNIST dataset (if this is your first time using this
# dataset then the 55MB download may take a minute)
print("[INFO] accessing MNIST...")
dataset = datasets.fetch_mldata("MNIST Original")
data = dataset.data
```

```python
# if we are using "channels first" ordering, then reshape the
# design matrix such that the matrix is:
# num_samples x depth x rows x columns
if K.image_data_format() == "channels_first":
    data = data.reshape(data.shape[0], 1, 28, 28)
# otherwise, we are using "channels last" ordering, so the design
# matrix shape should be: num_samples x rows x columns x depth
else:
    data = data.reshape(data.shape[0], 28, 28, 1)
```

```python
# scale the input data to the range [0, 1] and perform a train/test
# split
(trainX, testX, trainY, testY) = train_test_split(data / 255.0,
    dataset.target.astype("int"), test_size=0.25, random_state=42)
```

```python
# convert the labels from integers to vectors
le = LabelBinarizer()
trainY = le.fit_transform(trainY)
testY = le.transform(testY)
```

```python
# initialize the optimizer and model
print("[INFO] compiling model...")
opt = SGD(lr=0.01)
model = LeNet.build(width=28, height=28, depth=1, classes=10)
model.compile(loss="categorical_crossentropy", optimizer=opt, metrics=["accuracy"])
```

```python
# train the network
print("[INFO] training network...")
H = model.fit(trainX, trainY, validation_data=(testX, testY), batch_size=128, epochs=20, verbose=1)
```

```python
# evaluate the network
print("[INFO] evaluating network...")
predictions = model.predict(testX, batch_size=128)
c_report = classification_report(testY.argmax(axis=1), predictions.argmax(axis=1), target_names=[str(x) for x in le.classes_])
print(c_report)
```

```python
from ndl.utils.plot import plotClassificationReport
plotClassificationReport(c_report)
```

```python
from ndl.utils.plot import plotKerasReport
plotKerasReport(H, 20)
```

Here we see the "canonical" example of what we're looking for. Training and Validation accuracy mimic each other with no sights of overfitting.

It's rare that we see this in real-world datasets though. The MNIST dataset is heavily preprocessed to make it a benchmark. Any algorithm that can't be tuned to achieve > 95% on MNIST is probably badly implemented.
