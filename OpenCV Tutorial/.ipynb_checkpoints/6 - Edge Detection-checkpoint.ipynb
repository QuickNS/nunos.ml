```python
import cv2
from imutils import displayImage
import numpy as np
```

```python
image_path = 'images/coins.png'
```

```python
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image = cv2.GaussianBlur(image, (5, 5), 0)
displayImage(image, shiftColor=False)
```

## Laplacian Edge Detection

```python
# gradient calculation needs to be done on floats
lap = cv2.Laplacian(image, cv2.CV_64F)
# we then clip values by converting back to unsigned int 8 with the absolute value
# white-to-black transitions are negative, black-to-white are positive
lap = np.uint8(np.absolute(lap))
displayImage(lap, shiftColor=False)
```

## Sobel

```python
# Compute gradients along the X and Y axis, respectively
sobelX = cv2.Sobel(image, cv2.CV_64F, 1, 0)
sobelY = cv2.Sobel(image, cv2.CV_64F, 0, 1)

sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))

# We can combine our Sobel gradient images using our
# bitwise OR
sobelCombined = cv2.bitwise_or(sobelX, sobelY)

displayImage(sobelX, shiftColor=False, title='Sobel X')
displayImage(sobelY, shiftColor=False, title='Sobel Y')
displayImage(sobelCombined, shiftColor=False, title='Combined')
```

Edges are fairly noisy.

## Canny Edge Detection

```python
canny = cv2.Canny(image, 30, 150)
displayImage(canny, shiftColor=False)
```

```python
# values over third parameter are considered Edges
# values lower then second parameter are considered not-edges
# values in between depend on how intensities are connected
canny = cv2.Canny(image, 30, 250)
displayImage(canny, shiftColor=False)
```

```python
# we can further try to reduce noise
canny = cv2.Canny(image, 100, 250)
displayImage(canny, shiftColor=False)
```

```python
# changing blur can help as well
image = cv2.imread(image_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (11, 11), 0)
edged = cv2.Canny(blurred, 30, 150)
displayImage(edged, shiftColor=False)
```
