```python
import numpy as np
from imutils import displayImage
import cv2
```

```python
image = cv2.imread('images/trex.png')
displayImage(image)
```

## Blurring

Blurring is usually unwanted in our photographs but it's actually quite helpful while doing image processing. A lot of techniques such as edge detection and thresholding work better when the image is first blurred

## Averaging

```python
# this is applying a convolutional kernel as CNNs do
# applies a sliding window over a image where the pixel at the center becomes the average
blurred = np.vstack([
    cv2.blur(image, (3, 3)),
    cv2.blur(image, (5, 5)),
    cv2.blur(image, (7, 7))])
displayImage(blurred, figSize=(8,20))
```

## Gaussian

```python
# same principle but applying weighted mean where pixels closer contribute more. 
blurred = np.vstack([
    cv2.GaussianBlur(image, (3, 3), 0),
    cv2.GaussianBlur(image, (5, 5), 0),
    cv2.GaussianBlur(image, (7, 7), 0)])
displayImage(blurred, figSize=(8,20))
```

## Median
Very useful in removing 'salt and pepper' noise

```python
# same principle but applying median - which is always an actual pixel
blurred = np.vstack([
    cv2.medianBlur(image, 3),
    cv2.medianBlur(image, 5),
    cv2.medianBlur(image, 7)])
displayImage(blurred, figSize=(8,20))
```

## Bilateral
This reduces noise while preserving edges but it's slower to compute

```python
image = cv2.imread('images/beach.png')

blurred = np.vstack([
    cv2.bilateralFilter(image, 5, 21, 21),
    cv2.bilateralFilter(image, 7, 31, 31),
    cv2.bilateralFilter(image, 9, 41, 41)])
displayImage(blurred, figSize=(8,20))
```

Notice how noice is very reduced but edges are clearly retained (no motion blurring)
