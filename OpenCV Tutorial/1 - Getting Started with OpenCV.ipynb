```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
image_path = 'images/photo.jpg'
```

```python
def displayImage(image, shiftColor=True, figSize=(8,8)):
    fig = plt.figure(figsize=figSize)
    plt.axis("off")
    if shiftColor:
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    else:
        plt.imshow(image, cmap='Greys_r',  interpolation='bicubic')
    plt.show()
```

## Loading, Displaying and Saving

```python
image = cv2.imread(image_path)

# normally you would do this but it doesn't work very well in Jupyter notebooks
# cv2.imshow("Image", image)
# cv2.waitKey(0)

displayImage(image)
```

```python
# save and specify image format
cv2.imwrite("images/newimage.png", image)
```

## Accessing and Manipulating Pixels

```python
# Images are just NumPy arrays. The top-left pixel can be
# found at (0, 0)
(b, g, r) = image[0, 0]
print("Pixel at (0, 0) - Red: {}, Green: {}, Blue: {}".format(r, g, b))

# Now, let's change the value of the pixel at (0, 0) and
# make it red
image[0, 0] = (0, 0, 255)
(b, g, r) = image[0, 0]
print("Pixel at (0, 0) - Red: {}, Green: {}, Blue: {}".format(r, g, b))
```

```python
# Since we are using NumPy arrays, we can apply slicing and
# grab large chunks of the image. Let's grab the bottom-left
# corner - last 200 rows, first 200 columns 
corner = image[-200:, 0:200]
displayImage(corner, figSize=(2,2))
```

```python
# Let's make the top-left corner of the image green
# remember openCV uses BGR space as default
test = image
test[0:200, 0:200] = (0, 255, 0)
displayImage(test)
```

## Drawing

```python
# Initialize our canvas as a 300x300 with 3 channels,
# Red, Green, and Blue, with a black background
canvas = np.zeros((300, 300, 3), dtype = "uint8")

# Draw a green line from the top-left corner of our canvas
# to the bottom-right
green = (0, 255, 0)
cv2.line(canvas, (0, 0), (300, 300), green)
displayImage(canvas, figSize=(6,6))
```

```python
# Now, draw a 3 pixel thick red line from the top-right
# corner to the bottom-left
red = (0, 0, 255)
cv2.line(canvas, (300, 0), (0, 300), red, 3)
displayImage(canvas, figSize=(6,6))
```

```python
# Draw a green 50x50 pixel square, starting at 10x10 and
# ending at 60x60
cv2.rectangle(canvas, (10, 10), (60, 60), green)
displayImage(canvas, figSize=(6,6))
```

```python
# Draw another rectangle, this time we'll make it red and
# 5 pixels thick
cv2.rectangle(canvas, (50, 200), (200, 225), red, 5)
displayImage(canvas, figSize=(6,6))
```

```python
# Let's draw one last rectangle: blue and filled in
blue = (255, 0, 0)
cv2.rectangle(canvas, (200, 50), (225, 125), blue, -1)
displayImage(canvas, figSize=(6,6))
```

```python
# Reset our canvas and draw a white circle at the center
# of the canvas with increasing radii - from 25 pixels to
# 150 pixels
canvas = np.zeros((300, 300, 3), dtype = "uint8")
(centerX, centerY) = (canvas.shape[1] // 2, canvas.shape[0] // 2)
white = (255, 255, 255)

for r in range(0, 175, 25):
    cv2.circle(canvas, (centerX, centerY), r, white)
    
displayImage(canvas, figSize=(6,6))
```

```python
# Let's go crazy and draw 25 random circles
for i in range(0, 25):
    # randomly generate a radius size between 5 and 200,
    # generate a random color, and then pick a random
    # point on our canvas where the circle will be drawn
    radius = np.random.randint(5, high = 200)
    color = np.random.randint(0, high = 256, size = (3,)).tolist()
    pt = np.random.randint(0, high = 300, size = (2,))

    # draw our random circle
    cv2.circle(canvas, tuple(pt), radius, color, -1)  
    
displayImage(canvas, figSize=(6,6))
```
