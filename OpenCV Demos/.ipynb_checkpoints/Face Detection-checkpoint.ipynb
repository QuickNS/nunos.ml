```python
import cv2
import numpy as np
from utils.facedetector import FaceDetector
from utils.imutils import displayImage
```

```python
face_cascade = 'cascades/haarcascade_frontalface_default.xml'
image_path = 'images/obama2.png'
```

```python
image = cv2.imread(image_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
displayImage(gray)
```

```python
fd = FaceDetector(face_cascade)
faceRects = fd.detect(gray, scaleFactor=1.1, minNeighbors=4, minSize=( 30, 30))
print('I found {} face(s)'.format(len(faceRects)))

output = image.copy()
# draw rectangle for each face
for (x, y, w, h) in faceRects:
    cv2.rectangle(output, (x, y), (x+w, y+h), (0, 255, 0), 2)

displayImage(output)
```

```python
image_path = 'images/messi2.png'
image = cv2.imread(image_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

fd = FaceDetector(face_cascade)
faceRects = fd.detect(gray, scaleFactor=1.1, minNeighbors=4, minSize=( 30, 30))
print('I found {} face(s)'.format(len(faceRects)))

output = image.copy()
# draw rectangle for each face
for (x, y, w, h) in faceRects:
    cv2.rectangle(output, (x, y), (x+w, y+h), (0, 255, 0), 2)

displayImage(output)
```
